```python
deliveries_df.rename(columns={'total_runs_x':'total_runs','total_runs_y':'ball_score'},inplace=True)
```

```python
deliveries_df
```

```python
deliveries_df['Score'] = deliveries_df[['match_id','ball_score']].groupby('match_id').cumsum()['ball_score']
```

```python
deliveries_df[['ball_score','Score']]
```

```python
# New columns which has target 
deliveries_df['target_left'] = (deliveries_df['total_runs'] + 1) - deliveries_df['Score']
```

```python
deliveries_df
```

```python
# 1 over 6 balls 
# ipl has 20 over match 20 * 6 = 120 
deliveries_df['Remaining Balls'] = (120 - ( (deliveries_df['over'] - 1)*6 + deliveries_df['ball']))
```

```python
deliveries_df
```

```python
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].fillna('0') 
```

```python
deliveries_df
```

```python
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].apply(lambda x:x if x == '0' else '1').astype('int64')
```

```python
deliveries_df[['match_id','player_dismissed']]
```

```python
deliveries_df['Wickets'] = deliveries_df[['match_id','player_dismissed']].groupby('match_id').cumsum()['player_dismissed'].values
```

```python
deliveries_df['Wickets'] = 10 - deliveries_df['Wickets'] 
```

```python
deliveries_df
```

Current Run Rate 

$$Current \ run \ rate = \frac{Runs \ scored}{Number \ of \ Overs \ completed}$$

```python
deliveries_df['crr'] = (deliveries_df['Score'])*6/(120 - deliveries_df['Remaining Balls'])
```

```python
deliveries_df
```

Required Run Rate 

$$Required \ run \ rate = \frac{Remaining \ score \ or \ Target}{Number \ of \ Overs \ left}$$

```python
deliveries_df['rrr'] = (deliveries_df['target_left'])*6/(deliveries_df['Remaining Balls'])
```

```python
deliveries_df
```

```python
# Creating a column result where 1 means won 0 means lost 
def result(row):
    if row['batting_team'] == row['winner']: 
        return 1
    else: 
        return 0
```

```python
deliveries_df['result'] = deliveries_df.apply(result,axis=1)
```

```python
deliveries_df
```

```python
deliveries_df.columns
```

```python
Model_df = deliveries_df[['batting_team','bowling_team','city','Score','Wickets','Remaining Balls','target_left','crr','rrr','result']]
```

```python
Model_df.isnull().sum() 
```

```python
# Remove the records where city is missing 
Model_df[Model_df['city'].isna()]
```

```python
Model_df.dropna(inplace=True)
```

```python
Model_df.describe()
```

```python
Model_df = Model_df[Model_df['Remaining Balls'] != 0]
```

```python
Model_df = Model_df.sample(Model_df.shape[0])
```

```python
Model_df.sample()
```

```python
# Splitting the data into features and target 

X = Model_df.iloc[:,:-1]

Y = Model_df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
# Splitting the data to train and test : 
from sklearn.model_selection import train_test_split 

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
# Label encoding 
# One hot encoder 
# tranform non numberical data to numerical 

from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder 
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])], remainder = 'passthrough')
```

```python
trf
```

```python
# Building the model 
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_prediction = pipe.predict(X_test)
```

```python
Y_prediction
```

```python
# Accuracy score 
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_prediction)
```
